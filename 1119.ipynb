```python
#차대차 사건 중 죽거나 다친 사람이 많은 발생지 시도를 알아보고 시각화 해보자
#1. 사고유형_대분류_칼럼에서 차대차 사건만 인덱싱
#2. 1번 데이터에서 사상자수와 발생지시도 데이터를 인덱싱
#3. 2번 데이터에서 발생지시도로 그룹화하고 그룹화된 데이터의 합계를 구함.
#4. plot 그래프 그리기

#1번 정답
ctc = data[data['사고유형_대분류']=='차대차']
```

```python
#2번 정답
ctc_row=ctc[['사상자수','발생지시도']]
```

```python
#3번 정답
#groubby()
#sum()
a=ctc_row.groupby(['발생지시도']).sum()
a
```

```python
#4번 정답
x= a.index
y=a.values
```

```python
plt.figure(figsize=(15,8))
plt.plot(x,y)
plt.xlabel('지역')
plt.ylabel('사상자수')
plt.title('차 vs 차 교통사고의 사상자 수')
plt.show()
```
