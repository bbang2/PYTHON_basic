### train, test Data불러오기
- Passenger id 컬럼을 index

```python
import pandas as pd
```

```python
#컬럼을 인덱스로 가져오려면 index_col
train = pd.read_csv('train.csv', index_col = 'PassengerId')
test = pd.read_csv('test.csv', index_col = 'PassengerId')
```

```python
#데이터확인 info()
train.info()
```

```python
# 각 컬럼에 결측치가 몇 개 있나 합계 내보기
# null인가 아닌가 비교해주는 함수 : isnull()
# 합계를 내주는 함수 : sum() - 열(컬럼)단위
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
# Survived 0: 죽음, 1: 생존

# 생존한 사람 구분하기 : train['Survived']==1
# 주의 : train['Survived']=1 로 할 시 값이 다 1로 변함.

# 생존자의 데이터만 출력 : 구분한 값을 다시 train에 넣으면 생존자만 출력됨
train[train['Survived']==1]
```

```python
# 생존자 중 성별 칼럼 불러와서 항목 갯수 세기
# 성별 칼럼 불러오기 : ['Sex']
# 항목의 개수 세주기 : .value_counts()
Survived_1 = train[train['Survived']==1]['Sex'].value_counts()
```

```python
Survived_0 = train[train['Survived']==0]['Sex'].value_counts()
```

```python
# Survived_0, Survived_1 DataFrame 만들기
# index = Survived, Dead
# columns = Female, Male
df=pd.DataFrame([Survived_1, Survived_0], index=['Survived','Dead'])
```

```python
df
```

```python
# column 이름 변경 : rename
#rename(columns = {'현재 컬럼명':'바꿀 컬럼명'}, inplace=True)
df.rename(columns = {'female':'여자','male':'남자'}, inplace=True)
```

```python
df
```

```python
# 값 변환하기 : mapping
# 성별 칼럼 값 >> 여자:1  남자:0
```

```python
# 성별 칼럼 값이 남자인 데이터 불러오기
# 성별 칼럼 값이 여자인 데이터 불러오기
# loc 인덱서 사용해서 불러오기 : loc[행, 열]
train.loc[train['Sex']=='male','Sex']=0 # -> 이렇게 값을 변경하는게 매핑
# >>성별 남자인 (True)값을 보되, 성별 컬럼만 가져오겠다 = 값을 0으로 바꾸겠다
```

```python
train.loc[train['Sex']=='female','Sex']=1
```

```python
train['Sex'].value_counts()
```

```python
# Embarked 중 결측치가 존재하는 행 확인하기
# .isnull()함수 사용
# .notnull()함수 사용 시 결측치가 아닌 값을 불러옴
train[train['Embarked'].isnull()]
```

```python
# Embarked 특성의 데이터 갯수 세보기
train['Embarked'].value_counts()
```

```python
# Embarked 특성의 결측치를 'S'로 채우기
train=train.fillna('S')
```

```python
train[train['Embarked'].isnull()]
```

```python
# SibSp : 형제자매수
# Parch : 부모자식수
# FamilySize 특성 만들기 : (SibSp + Parch + 1)
```

```python
train['FamilySize']= train['SibSp']+train['Parch']+1
train['FamilySize']
```

```python
# 칼럼 맨앞으로 가져오기
del train['FamilySize']
```

```python
train = pd.read_csv('train.csv', index_col = 'PassengerId')
```

```python
# FamilySize 변수를 DataFrame으로 만들고
# FamilySize와 train을 concat함수를 통해서 합치기
FamilySize= train['SibSp']+train['Parch']+1
```

```python
FS=pd.DataFrame(FamilySize)
```

```python
FS.rename(columns={0:'FamilySize'},inplace=True)
# 컬럼 이름 변경
```

```python
final=pd.concat([FS,train],axis=1)
# FS를 뒤에다 쓰면 맨 뒤로 감
```
