# 카테고리

```python
ages=[0,2,10,21,23,37,31,61,20,41,32,100]

# 범위 >> 앞의 수 초과 뒤의 수 이하
bins=[0,15,25,35,60,99]

# 이름 붙여주기
labs=["미성년자","청년","중년","장년","노년"]
```

```python
#pd.cut(나눌데이터, 범위데이터, 라벨데이터)
cats=pd.cut(ages,bins,labels=labs)
cats
```

```python
cats[0]
```

```python
ageArr= pd.DataFrame(ages,columns=["나이"])
ageArr
```

## ageArr에 age_cat column 추가하기
### 카테고리로 나눈 데이터 추가

```python
ageArr["age_cat"]=cats
#=pd.cut(ages,bins,labels=labs)로 바로 받아도 됨
ageArr
```
