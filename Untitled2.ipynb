```python
x=[96,96,96,96,96,96,96]
y=[76,87,116,90,63,101,66]

plt.plot(x,label = '3~9세 평균 사용시간(분)')
plt.plot(y,label = '우리 아이 사용시간(분)')
plt.xlabel('사용기간(일)')
plt.ylabel('사용시간(분)')
plt.legend()
plt.show()
```

```python
x=np.arange(7)
a=[96,96,96,96,96,96,96]
day=['월','화','수','목','금','토','일']
real=[76,87,116,90,63,101,66]
plt.bar(x,real, width=0.6,color='peru',label='우리아이 스마트폰 사용시간(분)')
plt.plot(x,a, color='darkorange',label = '3~9세 권장사용시간(분)')
plt.title('일별 스마트폰 사용시간')
plt.xlabel('사용기간(요일)')
plt.ylabel('사용시간(분)')
plt.xticks(x,day)
plt.show()
```
