```python
data = np.loadtxt('ratings.txt', delimiter="::", dtype="int")
data
```

```python
print("배열의 크기 : ",data.shape)
print("배열의 차원 : ",data.ndim)
print("배열의 개수 : ",data.size)
```

```python
print(data[:,2].mean())
print(np.mean(data[:,2]))
```

```python
user_id=np.unique(data[:,0])
user_id
```

```python
for i in user_id:
    for j in range(1,6041):
        if i==j :
            a=data[j,2]
            print(a.mean())
```

```python
num=0
for i in user_id:
    num+=1
    if i==num :
        a=data[num,2]
        print(a.mean())
```

```python
list1=[]
for i in user_id:
    a= np.mean(data[data[:,0]==i,2])
    list1.append(a)


print(list1)

data
```
